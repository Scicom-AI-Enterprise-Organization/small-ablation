```python
from glob import glob
checkpoints = sorted(glob('ds3-qwen2.5-72b-lora-256/*'), key = lambda x: int(x.split('-')[-1]))
checkpoints
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel

model_name = "ramdisk/Qwen2.5-72B-Instruct"

for checkpoint in checkpoints:
    print(checkpoint)
    
    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        torch_dtype="auto",
        device_map="auto",
    )
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = PeftModel.from_pretrained(model, checkpoint)
    model = model.merge_and_unload()
    new_path = 'nfs/nfs/' + checkpoint.replace('/', '-')
    model.save_pretrained(new_path)
    tokenizer.save_pretrained(new_path)

    del model, tokenizer
```
