```python
import requests
import json
import re
import os
from tqdm import tqdm

def parse(text):
    m = re.search(r"[A-F].", text)
    if m:
        return m.group().replace('.', '')

headers = {
    'Content-Type': 'application/json',
    'Authorization': 'Bearer ',
}
```

```python
with open('MalayMMLU_0shot.json') as fopen:
    malaymmlu = json.load(fopen)
```

```python
import time

for i in tqdm(range(len(malaymmlu))):
    filename = f'malaymmlu/{i}.json'
    try:
        with open(filename) as fopen:
            d = json.load(fopen)
            continue
    except:
        pass
    
    q = malaymmlu[i]['prompt']
    prompt = f'Berikut adalah soalan aneka pilihan. Sila berikan jawapan sahaja.\n\n{q}\nJawapan:'
    
    json_data = {
        'model': 'ILMU-text',
        'messages': [
            {
                'role': 'user',
                'content': prompt,
            },
        ],
        'max_tokens': 5,
        'temperature': 0.7,
    }

    for _ in range(5):
        try:
            response = requests.post('https://api.ytlailabs.tech/v1/chat/completions', headers=headers, json=json_data)
            r = response.json()['choices'][0]['message']['content']
            with open(filename, 'w') as fopen:
                json.dump(r, fopen)
            break
        except:
            time.sleep(5.0)
```

```python
results = 0
for i in tqdm(range(len(malaymmlu))):
    filename = os.path.join('malaymmlu', f'{i}.json')
    with open(filename) as fopen:
        d = json.load(fopen)
    p = parse(d)
    if p:
        results += (p == malaymmlu[i]['key'])
results / len(malaymmlu)
```
