```python
from datasets import load_from_disk
ds = load_from_disk("./cleaned_dataset")
```

```python
ds[48]
```

```python
df.set_index("index", inplace=True)
```

```python
ds_map = ds.map(lambda x, i: {
    "response": df.loc[i]["response"],
    "response_lang": df.loc[i]["response_lang"]
}, with_indices=True)
```

```python
ds_map_filter = ds_map.filter(lambda x: x["response_lang"] != x["switch_language"] and x["response_lang"] != "unknown")
ds_map_filter.to_pandas().groupby("switch_language").size() / ds_map.to_pandas().groupby("switch_language").size()
```

```python
df_not_match = ds_map_filter.to_pandas()
```
