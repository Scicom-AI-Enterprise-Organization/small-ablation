# Push Datasets to HuggingFace

```python
import os
from huggingface_hub import HfApi

HF_TOKEN = os.environ.get('HF_TOKEN', 'your-token-here')
ORG = 'Scicom-intl'

DATASETS = {
    'turn-detector-Qwen3-0.6B-train-merged': '/root/small-ablation/call-center-language-switching/parquet-train-merged-v2',
    'turn-detector-Qwen3-0.6B-train': '/root/small-ablation/call-center-language-switching/parquet-train-v2',
    'turn-detector-Qwen3-0.6B-test': '/root/small-ablation/call-center-language-switching/parquet-test-v2',
}

api = HfApi(token=HF_TOKEN)
print('Ready')
```

```python
for name, path in DATASETS.items():
    repo_id = f'{ORG}/{name}'
    print(f'\n=== {repo_id} ===')
    print(f'  Source: {path}')
    
    api.create_repo(repo_id=repo_id, repo_type='dataset', exist_ok=True, private=False)
    
    api.upload_folder(
        folder_path=path,
        repo_id=repo_id,
        repo_type='dataset',
        commit_message=f'upload {name}',
    )
    print(f'  Done: https://huggingface.co/datasets/{repo_id}')

print('\nAll done!')
```
