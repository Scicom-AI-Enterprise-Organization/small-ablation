# Push Datasets to HuggingFace

```python
import os
from huggingface_hub import HfApi

HF_TOKEN = os.environ.get('HF_TOKEN', '')
REPO_ID = 'Scicom-intl/turn-detector-Qwen3-0.6B-dataset'

FOLDERS = {
    'train-merged': '/root/small-ablation/call-center-language-switching/parquet-train-merged-v2',
    'train': '/root/small-ablation/call-center-language-switching/parquet-train-v2',
    'test': '/root/small-ablation/call-center-language-switching/parquet-test-v2',
}

api = HfApi(token=HF_TOKEN)
print('Ready')
```

```python
api.create_repo(repo_id=REPO_ID, repo_type='dataset', exist_ok=True, private=False)

for folder_name, local_path in FOLDERS.items():
    print(f'\nUploading {folder_name} from {local_path}')
    api.upload_folder(
        folder_path=local_path,
        path_in_repo=folder_name,
        repo_id=REPO_ID,
        repo_type='dataset',
        commit_message=f'upload {folder_name}',
    )
    print(f'  Done')

print(f'\nAll done: https://huggingface.co/datasets/{REPO_ID}')
```
