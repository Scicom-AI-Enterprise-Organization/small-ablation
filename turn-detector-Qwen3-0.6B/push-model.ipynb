# Push Model to HuggingFace

```python
import os
from huggingface_hub import HfApi

CHECKPOINT_DIR = '/root/small-ablation/call-center-language-switching/output-0.6B-v2/checkpoint-1300'
REPO_ID = 'Scicom-intl/turn-detector-Qwen3-0.6B'
HF_TOKEN = os.environ.get('HF_TOKEN', 'your-token-here')

print(f'Checkpoint: {CHECKPOINT_DIR}')
print(f'Repo: {REPO_ID}')
```

```python
api = HfApi(token=HF_TOKEN)

api.create_repo(repo_id=REPO_ID, exist_ok=True, private=False)
print(f'Repo ready: https://huggingface.co/{REPO_ID}')
```

```python
api.upload_folder(
    folder_path=CHECKPOINT_DIR,
    repo_id=REPO_ID,
    commit_message='upload turn-detector-Qwen3-0.6B checkpoint-1300',
)
print(f'Done: https://huggingface.co/{REPO_ID}')
```
