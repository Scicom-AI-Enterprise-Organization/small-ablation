```python
import torch
import numpy as np
import math
```

```python
optimizer = torch.load(
    "malaysian-reasoning-20b-lora-r64-experts/checkpoint-100/optimizer.bin",
    map_location=torch.device("cpu"),
    weights_only=True,
)
```

```python
optimizer['state'][f'model.layers.0.mlp.experts.lora_gate_up_A.e.weight']['exp_avg_sq'].shape
```

```python
scores = {}
for i in range(24):
    tensor = optimizer['state'][f'model.layers.{i}.mlp.experts.lora_gate_up_A.e.weight']['exp_avg_sq']
    per_expert = tensor.mean(dim=1).sqrt()

    k = 4
    expert_score = per_expert.topk(k).values.mean()
    expert_score = expert_score / math.sqrt(tensor.numel())

    scores[i] = expert_score.item()

sorts = sorted(scores, key=scores.get, reverse=True)
for s in sorts:
    print(s, scores[s])
```
